```python
import sys
from pathlib import Path

PROJECT_ROOT = Path().resolve().parent
sys.path.insert(0, str(PROJECT_ROOT))
```

```python
from pathlib import Path
from src.indexer import (
    DatasetIndexer,
    RandomSplitter,
    save_jsonl,
    save_split_list,
)
```

```python
images_root = PROJECT_ROOT / "data" / "frames_raw"
labels_root = PROJECT_ROOT / "data" / "labels"
```

```python
indexer = DatasetIndexer(
    images_root=images_root,
    labels_root=labels_root,
    label_ext=".txt",
)
```

```python
items = indexer.build_index()
print(f"Total items found: {len(items)}")
```

```python
splitter = RandomSplitter(seed=42)
train, test = splitter.split_train_test(items, train_ratio=0.8)
print(f"Train items: {len(train)} | Test items: {len(test)}")
```

```python
save_jsonl(train + test, PROJECT_ROOT / "data/artifacts/dataset_index.jsonl")
save_split_list(train, PROJECT_ROOT / "data/artifacts/train_split.txt", use_rel_paths=True)
save_split_list(test, PROJECT_ROOT / "data/artifacts/test_split.txt", use_rel_paths=True)
```
