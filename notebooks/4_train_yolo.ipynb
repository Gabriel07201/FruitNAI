```python
from ultralytics import YOLO
```

```python
data_yaml = "../datasets/fruitninja_yolo/data.yaml"
```

```python
model = YOLO("yolo11n.pt")
```

```python
results = model.train(
    data=str(data_yaml),
    epochs=120,
    imgsz=640,
    batch=16,
    device=0,
    project="../models/runs",
    name="fruitninja_yolo11n",
)
```
